```python

import pandas as pd
import matplotlib.pyplot as plt
import mpl_toolkits.basemap
import numpy as np
import pylab as pl
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
#'D:/Nanohub/results/GIS1/EX2009core8.csv'
df = pd.DataFrame.from_csv('D:/Nanohub/results/GIS1/EX2009core8.csv', sep =',', encoding = 'UTF-8', header = None, index_col = False)
#df.columns = ['latitude','longitude','DOY','res_id','cluster_id']
```

```python
df.columns = ['latitude','longitude','DOY','res_id','cluster_id']
```

```python
col = df.groupby('cluster_id').count()
```

```python
xs = df['latitude']
ys = df['longitude']
zs = df['DOY']
c = df['cluster_id']
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(xs, ys, zs, c=c)
ax.set_xlabel('Latitude')
ax.set_ylabel('Longitude')
ax.set_zlabel('Day of the Year')
plt.show()
```

```python
#globe projection

fig = plt.figure(figsize=(20,10))

# Create a map, using the GallPeters projection, 
map = Basemap(projection='robin', 
              # with low resolution,
              resolution = 'l', 
              # And threshold 100000
              area_thresh = 100000.0,
              # Centered at 0,0 (i.e null island)
              lat_0=0, lon_0=0)

# Draw the coastlines on the map
map.drawcoastlines()

# Draw country borders on the map
map.drawcountries()

# Fill the land with grey
map.fillcontinents(color = '#888888')

# Draw the map boundaries
map.drawmapboundary(fill_color='#f4f4f4')

# Define our longitude and latitude points
# We have to use .values because of a wierd bug when passing pandas data
# to basemap.
x,y = map(df['longitude'].values, df['latitude'].values)

# Plot them using round markers of size 6
map.plot(x, y, 'bo', markersize=5)

# Show the map
plt.show()
```
