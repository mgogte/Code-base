```python
%matplotlib notebook
import pandas as pd
import csv
import numpy as np
import time 
import datetime
import matplotlib.pyplot as plt
import pylab as pl
```

```python
df = pd.DataFrame.from_csv('D:/Nanohub/TribeFlow/in_STDB_2015_5.csv', sep ='\t', encoding = 'UTF-8', header = None)
df.columns = ['ip','resource_id','datetime','latitude','longitude']
df['datetime'] = pd.to_datetime(df['datetime'])
date_col = []
for date in df['datetime']:
    date_col.append(date.date())
df['date_col'] = date_col
```

```python
#http://stackoverflow.com/questions/31569549/how-to-groupby-a-dataframe-in-pandas-and-keep-columns
df_timeline = df[['ip','date_col']].groupby('date_col').agg(['count']).reset_index()
```

```python
df_timeline.plot(x ='date_col', figsize = (60,30))
```
